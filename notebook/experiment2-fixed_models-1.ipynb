```python
import argparse

import os
import torch
import time
import pandas as pd
from tqdm import tqdm
import transformers
from transformers import AutoModelForCausalLM, AutoTokenizer, AutoModel
```

```python
import subprocess
import torch

def get_best_gpu():
    result = subprocess.check_output(
        [
            "nvidia-smi",
            "--query-gpu=memory.free",
            "--format=csv,noheader,nounits",
        ]
    )

    free_memories = [int(x) for x in result.decode().strip().split("\n")]
    return max(range(len(free_memories)), key=lambda i: free_memories[i])

gpu_id = get_best_gpu()
print("Best GPU:", gpu_id)

torch.cuda.set_device(gpu_id)
```

```python
class Base:
    def __init__(self, api_key='EMPTY'):
        self.name = None
        self.api_key = api_key

    def _get_response(self, inputs):
        raise NotImplementedError

    def get_response(self, src_file, key):
        directory_path, filename = os.path.split(src_file)
        tgt_file = os.path.join(directory_path, filename[:-4]+'_'+self.name+'_response.csv')

        df = pd.read_csv(src_file)
        tqdm.pandas()
        df[f'{self.name}_response'] = df[key].progress_apply(self._get_response)
        df.to_csv(tgt_file)
        return df
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM


class Llama3_8B(Base):
    def __init__(self, gpu_id=gpu_id):
        self.name = "llama3-8b-instruct"
        self.model_name_or_path = "meta-llama/Meta-Llama-3-8B-Instruct"

        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name_or_path)

        self.model = AutoModelForCausalLM.from_pretrained(
            self.model_name_or_path,
            torch_dtype=torch.bfloat16,
            device_map={"": gpu_id},
        )

        # LLaMA 3 uses EOS token properly
        self.eos_token_id = self.tokenizer.eos_token_id

    def _get_response(self, inputs):

        # 1. build chat format (IMPORTANT)
        messages = [{"role": "user", "content": inputs}]

        text = self.tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True,
        )

        # 2. tokenize
        model_inputs = self.tokenizer(
            text,
            return_tensors="pt",
            truncation=True
        ).to(self.model.device)

        # 3. generate 
        outputs = self.model.generate(
            **model_inputs,
            max_new_tokens=256,
            do_sample=True,
            temperature=0.6,
            top_p=0.9,
            repetition_penalty=1.1,
            eos_token_id=self.eos_token_id,
        )

        # 4. extract only new tokens
        gen_ids = outputs[:, model_inputs.input_ids.shape[1]:]

        return self.tokenizer.decode(
            gen_ids[0],
            skip_special_tokens=True
        ).strip()
```

```python
from langdetect import detect

from deep_translator import GoogleTranslator

def translate_to_english(text):
    return GoogleTranslator(source='auto', target='en').translate(text)
```

```python
l4 = pd.read_csv("l4_balanced_2.csv")
```

```python
from tqdm import tqdm
from langdetect import detect

responses = []

for i, prompt in enumerate(tqdm(l4["prompt"])):
    try:
        response = model._get_response(prompt)

        if response:
            try:
                lang = detect(response)
            except:
                lang = "unknown"

            if lang != "en":
                try:
                    response = translate_to_english(response)
                except Exception as e:
                    print(f"Translation error at row {i}: {e}")

        else:
            response = None

    except Exception as e:
        print(f"Error at row {i}: {e}")
        response = None

    responses.append(response)

    if (i + 1) % 10 == 0:
        l4.loc[l4.index[:len(responses)], "response"] = responses
        l4.to_csv("llama3/l4_balance_with_response.csv", index=False)

l4["response"] = responses
l4.to_csv("llama3/l4_balance_with_response.csv", index=False)
```
