```python
import subprocess
import torch

def get_best_gpu():
    result = subprocess.check_output(
        [
            "nvidia-smi",
            "--query-gpu=memory.free",
            "--format=csv,noheader,nounits",
        ]
    )

    free_memories = [int(x) for x in result.decode().strip().split("\n")]
    return max(range(len(free_memories)), key=lambda i: free_memories[i])

gpu_id = get_best_gpu()
print("Best GPU:", gpu_id)

torch.cuda.set_device(gpu_id)
```

```python
class Base:
    def __init__(self, api_key='EMPTY'):
        self.name = None
        self.api_key = api_key

    def _get_response(self, inputs):
        raise NotImplementedError

    def get_response(self, src_file, key):
        directory_path, filename = os.path.split(src_file)
        tgt_file = os.path.join(directory_path, filename[:-4]+'_'+self.name+'_response.csv')

        df = pd.read_csv(src_file)
        tqdm.pandas()
        df[f'{self.name}_response'] = df[key].progress_apply(self._get_response)
        df.to_csv(tgt_file)
        return df
```

```python
import torch
from transformers import AutoTokenizer, AutoModel


class ChatGLM3(Base):
    def __init__(self, api_key="EMPTY", gpu_id=gpu_id):
        self.name = "chatglm3-6b-base"

        self.model_name_or_path = "THUDM/chatglm3-6b" #"zai-org/chatglm3-6b-base"

        self.tokenizer = AutoTokenizer.from_pretrained(
            self.model_name_or_path,
            trust_remote_code=True,
        )

        self.model = AutoModel.from_pretrained(
            self.model_name_or_path,
            trust_remote_code=True,
            torch_dtype=torch.float16,
        ).cuda(gpu_id)

        self.model = self.model.eval()

    # def _get_response(self, inputs):
    #     response, history = self.model.chat(
    #         self.tokenizer,
    #         inputs,
    #         history=[],
    #     )

    #     return response

    # def _get_response(self, prompt):

    #     inputs = self.tokenizer(
    #         prompt,
    #         return_tensors="pt"
    #     ).to(self.model.device)

    #     outputs = self.model.generate(
    #         **inputs,
    #         max_new_tokens=512,
    #         do_sample=False,
    #     )

    #     response = self.tokenizer.decode(
    #         outputs[0][inputs["input_ids"].shape[1]:],
    #         skip_special_tokens=True,
    #     )

    #     return response

    def _get_response(self, prompt):

        messages = [
            {
                "role": "system",
                "content": "Answer ONLY in English. If the question is written in Chinese, or any other language, translate it internally and answer in English. Question:"
            },
            {
                "role": "user",
                "content": prompt
            }
        ]

        # inputs = self.tokenizer.apply_chat_template(
        #     messages,
        #     tokenize=True,
        #     return_tensors="pt",
        #     return_dict=True,
        #     add_generation_prompt=True,
        # ).to(self.model.device)

        inputs = self.tokenizer(
            prompt,
            return_tensors="pt"
        ).to(self.model.device)


        outputs = self.model.generate(
            **inputs,
            max_new_tokens=512,
            do_sample=False,
        )

        response = self.tokenizer.decode(
            outputs[0][inputs["input_ids"].shape[1]:],
            skip_special_tokens=True,
        )

        return response.strip()
```

```python
from tqdm import tqdm
from langdetect import detect

responses = []

for i, prompt in enumerate(tqdm(l1["prompt"])):
    try:
        response = model._get_response(prompt)

        if response:
            try:
                lang = detect(response)
            except:
                lang = "unknown"

            if lang != "en":
                try:
                    response = translate_to_english(response)
                except Exception as e:
                    print(f"Translation error at row {i}: {e}")

        else:
            response = None

    except Exception as e:
        print(f"Error at row {i}: {e}")
        response = None

    responses.append(response)

    if (i + 1) % 10 == 0:
        l1.loc[l1.index[:len(responses)], "response"] = responses
        l1.to_csv("chatglm3/l1_balance_with_response.csv", index=False)

l1["response"] = responses
l1.to_csv("chatglm3/l1_balance_with_response.csv", index=False)
```

```python
import pandas as pd
from tqdm import tqdm
from langdetect import detect

l4 = pd.read_csv("chatglm3/l4_balance_with_response.csv")

# các dòng chưa có response
missing_idx = l4[
    l4["response"].isna() |
    (l4["response"].astype(str).str.strip() == "")
].index

print(f"Missing rows: {len(missing_idx)}")

for idx in tqdm(missing_idx):
    prompt = l4.at[idx, "prompt"]

    try:
        response = model._get_response(prompt)

        if response and response.startswith(prompt):
            response = response[len(prompt):].strip()

        if response:
            try:
                lang = detect(response)
            except:
                lang = "unknown"

            if lang != "en":
                try:
                    response = translate_to_english(response)
                except Exception as e:
                    print(f"Translation error at row {idx}: {e}")

        l4.at[idx, "response"] = response

    except Exception as e:
        print(f"Error at row {idx}: {e}")

    # save every 10 completed rows
    if (list(missing_idx).index(idx) + 1) % 10 == 0:
        l4.to_csv(
            "chatglm3/l4_balance_with_response.csv",
            index=False
        )

l4.to_csv(
    "chatglm3/l4_balance_with_response.csv",
    index=False
)
```
